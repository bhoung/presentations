```python
text = open(filename, 'r').read()
soup = BeautifulSoup(text, 'html.parser')
#print(soup)
```

```python
tables = soup.find_all('table', attrs={'class':'zebra player'})
```

```python
table = tables[0]
```

```python
table
```

```python
name = table.attrs['id']
rows = table.find_all('tr')
list = []

row = rows[0]
row
```

```python
with open("bl-v-wb.csv", "wt+", newline="") as f:
   writer = csv.writer(f)
   for row in rows:    
      csv_row = []
      for cell in row.findAll(["td", "th"]):
          name = cell.find('span', attrs={'class':'full-name'})
          try:
              name = name.text.replace('\xa0', ' ')
              csv_row.append(name)
          except:
              txt = cell.text.strip()
              txt = txt.replace('\\n', '')
              txt = txt.replace('\\t', '')
              csv_row.append(txt)
              #csv_row.append(cell.text.strip())
      print(csv_row)
      writer.writerow(csv_row)

   
```
